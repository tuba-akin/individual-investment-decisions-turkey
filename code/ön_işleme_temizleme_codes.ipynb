```python
import pandas as pd
```

```python
# tufe_raw.csv ön işleme

tufe = pd.read_csv("tufe_raw.csv", sep=";", encoding="utf-8-sig", decimal=",")

tufe["Tarih"] = pd.to_datetime(tufe["Tarih"], format="%Y-%m")
tufe["TÜFE_Yıllık_Değişim"] = pd.to_numeric(tufe["TÜFE_Yıllık_Değişim"], errors="coerce")
tufe["TÜFE_Aylık_Değişim"] = pd.to_numeric(tufe["TÜFE_Aylık_Değişim"], errors="coerce")

tufe = tufe[["Tarih", "TÜFE_Yıllık_Değişim", "TÜFE_Aylık_Değişim"]]
tufe = tufe[tufe["Tarih"] >= "2021-01-01"]

tufe.head()
```

```python
# politika_faiz_raw.csv ön işleme

politika = pd.read_csv("politika_faiz_raw.csv", sep=";", encoding="utf-8-sig", decimal=",")

politika["Tarih"] = pd.to_datetime(politika["Tarih"], format="%Y-%m")
politika["Politika_Faizi"] = pd.to_numeric(politika["POLFAIZ_TUR"], errors="coerce")

politika = politika[["Tarih", "Politika_Faizi"]]
politika = politika[politika["Tarih"] >= "2021-01-01"]

politika.head()
```

```python
# mevduat_faiz_raw.csv ön işleme

mevduat = pd.read_csv("mevduat_faiz_raw.csv", sep=";", encoding="utf-8-sig", decimal=",")

mevduat["Tarih"] = pd.to_datetime(mevduat["Tarih"], format="%d.%m.%Y")
mevduat["Mevduat_Faizi"] = pd.to_numeric(mevduat["Mevduat_Yüzde"], errors="coerce")

mevduat = mevduat[["Tarih", "Mevduat_Faizi"]]
mevduat = mevduat[mevduat["Tarih"] >= "2021-01-01"]

# haftalık veriyi aylık ortalamaya çevirme
mevduat["Tarih"] = mevduat["Tarih"].values.astype("datetime64[M]")
mevduat = mevduat.groupby("Tarih", as_index=False)["Mevduat_Faizi"].mean()

# virgülden sonra 2 basamak olacak şekilde yuvarlama
mevduat["Mevduat_Faizi"] = mevduat["Mevduat_Faizi"].round(2)

mevduat.head()
```

```python
# dolar_kuru_raw.csv ön işleme

dolar = pd.read_csv("dolar_kuru_raw.csv", sep=";", encoding="utf-8-sig", decimal=",")

dolar["Tarih"] = pd.to_datetime(dolar["Tarih"], format="%Y-%m")
dolar["USDTRY"] = pd.to_numeric(dolar["USD_YTL"], errors="coerce")

dolar = dolar[["Tarih", "USDTRY"]]
dolar = dolar[dolar["Tarih"] >= "2021-01-01"]

dolar.head()
```

```python
# altin_raw.csv ön işleme
altin = pd.read_csv("altın_raw.csv", sep=";", encoding="utf-8-sig", decimal=",")
altin["Tarih"] = pd.to_datetime(altin["Tarih"], format="%Y-%m")

# altın fiyatını temizleyip sayıya çevirme
altin["Gram_Altın"] = (
    altin["ALTIN_YTL"]
    .astype(str)
    .str.replace(".", "", regex=False)   # binlik ayırıcıyı kaldır
    .str.replace(",", ".", regex=False)  # ondalık ayırıcıyı düzelt
    .pipe(pd.to_numeric, errors="coerce")
)

# sadece gerekli sütunlar
altin = altin[["Tarih", "Gram_Altın"]]
altin = altin[altin["Tarih"] >= "2021-01-01"]

altin.tail()
```

```python
# bist100_raw.csv ön işleme

bist = pd.read_csv("bist100_raw.csv", encoding="utf-8-sig")

bist["Tarih"] = pd.to_datetime(bist["Tarih"], format="%d.%m.%Y")
bist["BIST100"] = (
    bist["Şimdi"]
    .astype(str)
    .str.replace(".", "", regex=False)
    .str.replace(",", ".", regex=False)
    .pipe(pd.to_numeric, errors="coerce")
)

bist = bist[["Tarih", "BIST100"]]
bist = bist[bist["Tarih"] >= "2021-01-01"]

bist.head()
```

```python
# brent_petrol_raw.csv ön işleme

brent = pd.read_csv("brent_petrol_raw.csv", encoding="utf-8-sig")

brent["Tarih"] = pd.to_datetime(brent["Tarih"], format="%d.%m.%Y")
brent["Brent_Petrol"] = (
    brent["Şimdi"]
    .astype(str)
    .str.replace(".", "", regex=False)
    .str.replace(",", ".", regex=False)
    .pipe(pd.to_numeric, errors="coerce")
)

brent = brent[["Tarih", "Brent_Petrol"]]
brent = brent[brent["Tarih"] >= "2021-01-01"]

brent.head()
```

```python
# bist_gida_raw.csv ön işleme

bist_gida = pd.read_csv("bist_gida_raw.csv", encoding="utf-8-sig")

bist_gida["Tarih"] = pd.to_datetime(bist_gida["Tarih"], format="%d.%m.%Y")

bist_gida["BIST_Gıda"] = (
    bist_gida["Şimdi"]
    .astype(str)
    .str.replace(".", "", regex=False)
    .str.replace(",", ".", regex=False)
    .pipe(pd.to_numeric, errors="coerce")
)

bist_gida = bist_gida[["Tarih", "BIST_Gıda"]]
bist_gida = bist_gida[bist_gida["Tarih"] >= "2021-01-01"]

bist_gida.head()
```

```python
# bist_kimya_raw.csv ön işleme

bist_kimya = pd.read_csv("bist_kimya_raw.csv", encoding="utf-8-sig")

bist_kimya["Tarih"] = pd.to_datetime(bist_kimya["Tarih"], format="%d.%m.%Y")

bist_kimya["BIST_Kimya"] = (
    bist_kimya["Şimdi"]
    .astype(str)
    .str.replace(".", "", regex=False)
    .str.replace(",", ".", regex=False)
    .pipe(pd.to_numeric, errors="coerce")
)

bist_kimya = bist_kimya[["Tarih", "BIST_Kimya"]]
bist_kimya = bist_kimya[bist_kimya["Tarih"] >= "2021-01-01"]

bist_kimya.head()
```

```python
# bist_ulastirma_raw.csv ön işleme

bist_ulastirma = pd.read_csv("bist_ulastirma_raw.csv", encoding="utf-8-sig")

bist_ulastirma["Tarih"] = pd.to_datetime(bist_ulastirma["Tarih"], format="%d.%m.%Y")

bist_ulastirma["BIST_Ulaştırma"] = (
    bist_ulastirma["Şimdi"]
    .astype(str)
    .str.replace(".", "", regex=False)
    .str.replace(",", ".", regex=False)
    .pipe(pd.to_numeric, errors="coerce")
)

bist_ulastirma = bist_ulastirma[["Tarih", "BIST_Ulaştırma"]]
bist_ulastirma = bist_ulastirma[bist_ulastirma["Tarih"] >= "2021-01-01"]

bist_ulastirma.head()
```

```python
# tahvil_5_yillik_raw.csv ön işleme

tahvil_5y = pd.read_csv("tahvil_5_yillik_raw.csv", encoding="utf-8-sig")

tahvil_5y["Tarih"] = pd.to_datetime(tahvil_5y["Tarih"], format="%d.%m.%Y")

tahvil_5y["Tahvil_5Y_Faizi"] = (
    tahvil_5y["Şimdi"]
    .astype(str)
    .str.replace(".", "", regex=False)
    .str.replace(",", ".", regex=False)
    .pipe(pd.to_numeric, errors="coerce")
)

tahvil_5y = tahvil_5y[["Tarih", "Tahvil_5Y_Faizi"]]
tahvil_5y = tahvil_5y[tahvil_5y["Tarih"] >= "2021-01-01"]

tahvil_5y.head()
```

```python
# tüm verileri birleştirme

tarih_iskeleti = pd.DataFrame({
    "Tarih": pd.date_range("2021-01-01", "2026-03-01", freq="MS")
})

birlesik_veri = (
    tarih_iskeleti
    .merge(tufe, on="Tarih", how="left")
    .merge(politika, on="Tarih", how="left")
    .merge(mevduat, on="Tarih", how="left")
    .merge(dolar, on="Tarih", how="left")
    .merge(altin, on="Tarih", how="left")
    .merge(bist, on="Tarih", how="left")
    .merge(brent, on="Tarih", how="left")
    .merge(bist_gida, on="Tarih", how="left")
    .merge(bist_kimya, on="Tarih", how="left")
    .merge(bist_ulastirma, on="Tarih", how="left")
    .merge(tahvil_5y, on="Tarih", how="left")
)

birlesik_veri.head()
```

```python
# birleşik veri kontrolü

print(birlesik_veri.shape)
print(birlesik_veri.isna().sum())

birlesik_veri.head()
birlesik_veri.tail()
```

```python
birlesik_veri = birlesik_veri.dropna().reset_index(drop=True)
```
